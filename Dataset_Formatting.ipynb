Import all libraries to format datasets

```python
import pandas as pd
import os
```

Here is defined columns with actual data and indexes

```python
cols = ['2010', '2011', '2012', '2013', '2014','2015','2016','2017']
indeces = ['2010i', '2011i', '2012i', '2013i', '2014i', '2015i', '2016i']
```

Function for indeces calculation

```python
def calc_indeces(df):
    for i in range(len(indeces)):
        df[indeces[i]] = round(((df[cols[i + 1]] / df[cols[i]]) - 1) * 100, 2)
    df = df[['TIME']+indeces]
    return df
```

Function for transforming table into dataset with indeces

```python
def format_dataset(path):
    df = pd.read_csv(path+'/prepared.csv')
    df = calc_indeces(df)
    df.to_csv(path+'/formatted.csv',index=False)
```

Here is defined all paths with prepared tables

```python
tables = [x[0] for x in os.walk('data')]
##removing data folder from list
tables.pop(0)
```

Applying formatting and indeces calculation for all tables in folder table

```python
for table in tables:
    format_dataset(table)
```
